```python
# Merge,Join and Concatenate in Pandas
```

```python
import pandas as pd
import numpy as np
data1 = {"Emp Id":["E01","E02","E03","E04","E05","E06"],"Names":["Ram","Rahul","Jay","Ravi","Atul","Visal"],"Age":[56,34,23,45,24,34]}

data2 = {"Emp Id":["E01","E07","E03","E04","E08","E06"],"Salary":[56000,45000,40000,34000,44000,35000]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
print(df1)
print()
print(df2)
```

```python
print(pd.merge(df1,df2,on = "Emp Id"))
```

```python
print(pd.merge(left = df1,right = df2,on = "Emp Id",how = "left"))
```

```python
print(pd.merge(left = df1,right = df2,on = "Emp Id",how = "right"))
```

```python
# Pivoting and Melting in Data Frame
```

```python
import pandas as pd

dict = {"Keys":["K1","K2","K1","K2"],"Names":["Ram","Ben","Rahul","Peter"],"Houses":["red","blue","green","red"],"Grades":["3rd","8th","9th","8th"]}
df = pd.DataFrame(dict)

print(df)
```

```python
print(df.pivot(index = "Keys",columns = "Names",values = ["Houses","Grades"]))
```

```python
dict = {"Names":["Ram","Ben","Rahul","Peter"],"Houses":["red","blue","green","red"],"Grades":["3rd","8th","9th","8th"]}
df = pd.DataFrame(dict)
print (df)
```

```python
print(pd.melt(df,id_vars=["Names"],value_vars=["Houses","Grades"]))
```

```python
print(pd.melt(df,id_vars=["Names"],value_vars=["Houses","Grades"], var_name = "Houses&Grades",value_name = "Values"))
```
